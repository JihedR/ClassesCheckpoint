```python
#Question 1
class CompteBancaire:
    def __init__(self, numeroCompte, nom, sold):
        self.numeroCompte=numeroCompte
        self.nom = nom
        self.sold = sold
        
    def Versement(self, vers):
        self.sold += vers
        return "{} a ajouter {}. Son neauvau sold est: {}".format(self.nom,vers,self.sold)
    def Retrait(self, ret):
        self.sold -= ret
        return "{} a retraite {}. Son nouveau sold est: {}".format(self.nom,ret,self.sold)
    def Afficher(self):
        return "Le sold de {} avec le numero {} est {}".format(self.nom,self.numeroCompte,self.sold)

N1=CompteBancaire(123,"Jihed",1000)
print(N1.Versement(500))
print(N1.Retrait(250))
print(N1.Afficher())
```
